```python
# Задание 3
# окружность, эллипс,гипербола.
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axis()
plt.gca().set_aspect('equal')
# plt.axis([-6, 6, -6, 6])

x_ =np.linspace(-5,5,10000)

a=2
x0=0
y0=0
y_c = (a**2 - (x_-x0)**2)**0.5 - y0
y_c2 = -(a**2 - (x_-x0)**2)**0.5 - y0
plt.plot(x_,y_c,"b", label="circle")
plt.plot(x_,y_c2,"b")

b=1
y_e = ((1-(x_**2/a**2))*b**2)**0.5
plt.plot(x_,y_e,"r", label="ellipse")
plt.plot(x_,-y_e,"r")

y_h = (-(1-(x_**2/a**2))*b**2)**0.5
plt.plot(x_,y_h, "g",label="hyperbola")
plt.plot(x_,-y_h, "g")

plt.legend(loc = "best")
```

```python
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')

a=2
 
xx=[]
yy=[]
yy2=[]
i=-a
while i<a:
    xx.append(i)
    yy.append(math.sqrt(a**2 - i**2))
    yy2.append(-math.sqrt(a**2 - i**2))
    i=i+0.005
plt.plot(xx,yy)
plt.plot(xx,yy2)
```

```python
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')

a=3
b=1
xx=[]
yy=[]
yy2=[]
i=-a
while i<a:
    xx.append(i)
    yy.append(math.sqrt((1-(i**2/a**2))*b**2))
    yy2.append(-math.sqrt((1-(i**2/a**2))*b**2))
    i=i+0.005
plt.plot(xx,yy)
plt.plot(xx,yy2)
```

```python
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')

a=3
b=2
xx=[]
yy=[]
yy2=[]
i=-a-b
while i<a+b:
    xx.append(i)
    yy.append((((i**2/a**2)-1)*b**2)**0.5)
    yy2.append(-(((i**2/a**2)-1)*b**2)**0.5)
    i=i+0.01
plt.plot(xx,yy)
plt.plot(xx,yy2)
```
