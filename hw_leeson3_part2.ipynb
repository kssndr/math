```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
# plt.plot(x, np.tan(x))

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-1,1,201)
plt.plot(x, np.arcsin(x))
plt.plot(x, np.arccos(x))


i = [0.5,6,4,3,2,1]
plt.plot([-1,1],[np.pi/i[0],np.pi/i[0]], color="blue", linestyle ="--")
plt.plot([-1,1],[np.pi/i[1],np.pi/i[1]], color="blue", linestyle ="--")
plt.plot([-1,1],[np.pi/i[2],np.pi/i[2]], color="blue", linestyle ="--")
plt.plot([-1,1],[np.pi/i[3],np.pi/i[3]], color="blue", linestyle ="--")
plt.plot([-1,1],[np.pi/i[4],np.pi/i[4]], color="blue", linestyle ="--")
plt.plot([-1,1],[np.pi/i[5],np.pi/i[5]], color="blue", linestyle ="--")

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-1,1)
plt.ylim(-np.pi,np.pi)
plt.grid(True)
plt.show()
```

```python
for i in (0.5,6,4,3,2,1):
    print(f"cos pi/{i} = {np.cos(np.pi/i):.2f}  sin pi/{i} = {np.sin(np.pi/i):.2f} tang pi/{i} = {np.tan(np.pi/i):.2f}")
```

```python
# Задание 1

plt.axis([-5, 5, -5, 5])
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)


k = [4,-2,0.5]
a = [1,-2,0.5]
b = [0,-2,0.5]

x=np.linspace(-2*np.pi, 2*np.pi, 201)
y0=k[0]*np.cos(x-a[0])+b[0]
y1=k[1]*np.cos(x-a[1])+b[1]
y2=k[2]*np.cos(x-a[2])+b[2]


plt.plot(x,y0,label="0")
plt.plot(x,y1,label="1")
plt.plot(x,y2,label="2")

plt.legend()
```
