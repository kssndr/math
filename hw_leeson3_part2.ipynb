```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (10,7)

plt.ylim(-4,4)
plt.xlim(-2*np.pi,2*np.pi)

x = np.linspace(-2*np.pi, 2*np.pi, 201)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))

plt.plot([np.pi,np.pi],[-4,4], color="blue", linestyle ="--")
plt.plot([np.pi/2,np.pi/2],[-4,4], color="blue", linestyle ="--")
plt.plot([2*np.pi,2*np.pi],[-4,4], color="blue", linestyle ="--")
plt.plot([np.pi*3/2,np.pi*3/2],[-4,4], color="blue", linestyle ="--")
plt.plot([np.pi/2,np.pi/2],[-4,4], color="blue", linestyle ="--")
plt.plot([-np.pi,-np.pi],[-4,4], color="blue", linestyle ="--")
plt.plot([-np.pi/-2,np.pi/2],[-4,4], color="blue", linestyle ="--")
plt.plot([-2*np.pi,-2*np.pi],[-4,4], color="blue", linestyle ="--")
plt.plot([-np.pi*3/2,-np.pi*3/2],[-4,4], color="blue", linestyle ="--")
plt.plot([-np.pi/2,-np.pi/2],[-4,4], color="blue", linestyle ="--")
plt.plot([0,0],[-4,4], color="blue", linestyle ="--")

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')

plt.show()
```
