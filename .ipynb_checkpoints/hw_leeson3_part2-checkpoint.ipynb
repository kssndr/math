```python
plt.axis([-1, 3, -2, 2])
plt.gca().set_aspect('equal')
plt.rcParams["figure.figsize"] = (5,5)

x = np.linspace(-1,3,200)

# рассчет коэффициентов для полинома по заданным корням
k=np.poly([1,1,3])
print(k)

# рассчет корней полином по заданным коэффициентам
print(np.roots(k))

# вычесление значениея полинома по коэфф в точке х
plt.plot(x,np.polyval(k,x), "r:",label="b * x**3 + c*x**2 + d*x +e")

# a = k[0]
b = k[0]
c = k[1]
d = k[2]
e = k[3]
y = b * x**3 + c*x**2 + d*x +e
# y = a * x**4 + b * x**3 + c*x**2 + d*x +e


plt.plot(x,x**2,label="x**2")
plt.plot(x,np.exp(x),label="e")
# знак функции
plt.plot(x,np.sign(x))
plt.plot(x,y,label="b * x**3 + c*x**2 + d*x +e")


plt.grid(True)
plt.legend(loc = "best")
plt.show()

print(np.roots(k))
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve

plt.gca().set_aspect('equal')
plt.rcParams["figure.figsize"] = (10,10)
plt.axis([-5, 15, -2.5, 17.5])
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)

def f(x):
    return (x**2-1)

x1 = fsolve(f,-10)
print(x1)

def ec(p):
    x,y = p
    return(x**2 - 1 -y,np.exp(x)+ x - x*y - 1)

x2,y2 = fsolve(ec,(-2.5,-2.5))
print(x2,y2)


x1 = np.linspace(-10,10,200)
y1 = x1**2-1

x2 = np.linspace(-10,10,200)
y2 = (np.exp(x2)+ x2 - 1)/x2

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.scatter(4.200105840672318,16.640889072927425, color ="red", label = "root 3")
plt.scatter(2.618145573084665, 5.854686241862589, color ="red", label = "root 2")
plt.scatter(-1.5818353527550308, 1.502203082916442, color ="red", label = "root 1")
plt.legend(loc = "best")
```

Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y) > 1

2 корня: 4.200105840672318,16.640889072927425  и 2.618145573084665, 5.854686241862589
удовлетворяют систему неравенств (больше 1 и принадлежат y = x2 – 1)
